```python
mp_drawing = mp.solutions.drawing_utils
# mp_drawing_styles = mp.solutions.drawing_styles
mp_hands = mp.solutions.hands  # Solução para detecção e rastreio de mão e dedos
```

```python
# --- Variáveis globais ---
path = os.getcwd()
input_images_path = path + '\\images\\'
output_images_path = path + '\\output_images\\'
```

As duas saídas (outputs) desse modelo são:

### I. Multi_handedness

Coleção das mãos detectadas (ou seja, é uma mão esquerda ou direita). Cada mão é composta por um __label__ e um __score__. O label é uma string com o valor "Esquerda" ou "Direita". O score é a probabilidade estimada da mão prevista e é sempre maior ou igual a 0.5 (e a mão oposta tem uma probabilidade estimada de 1-score).

**Observação:** Note-se que a mão é determinada assumindo que a imagem de entrada é espelhada, ou seja, tirada com uma câmara virada para a frente com as imagens viradas na horizontal. Se não for este o caso, por favor trocar a saída da mão na aplicação.

---

### II. Multi_hand_landmarks

Conjunto de mãos detectadas, onde cada mão é representada como uma __lista de 21 marcos (pontos) de mão__ e cada marco de referência (ponto de referência) é composto por __x, y e z__. Desses, x e y são normalizados para [0.0, 1.0] pela largura e altura da imagem, respectivamente. Já z representa a profundidade do marco de referência, sendo a profundidade no pulso a origem (referência). Quanto menor o valor de z, mais próximo o marco de referência está da câmera. A magnitude de z usa aproximadamente a mesma escala que x.

### a. Vídeo

O bloco abaixo aciona a câmera webcam do notebook e aplica a detecção e rastreio das mãos e dedos na imagem. No pulso da mão detectada, aparece o label identifica ("left" ou  "right"), a precisão da identificação ("score") e as coordenadas desse ponto do pulso ("coords").

**Observação**: na imagem, o referencial (0,0) é o ponto no campo superior esquerda da imagem.

```python
def display_webcam(filename=None):
    """
    Inicializa webcam. Para parar o processo pressionar o botão 'q' e para salvar o frame (imagem) na pasta "images"
    pressione "s".
    """
    
    # Create a video capture object for the camera
    cap = cv2.VideoCapture(0)
    
    while(True):
        # Capture the video frame-by-frame
        _, frame = cap.read()

        # Display the resulting frame
        cv2.imshow('Frame', frame)
        
        # Press 'q' button to quit and 's' button to save the image
        if cv2.waitKey(1) & 0xFF == ord('q'):
            print("Terminando processo.")
            break
        elif cv2.waitKey(1) & 0xFF == ord('s'):
            if filename is None:
                print("Não foi passado nome de arquivo")
            else:
                try:
                    print("Salvar imagem")
                    cv2.imwrite(input_images_path + '\\' + filename, frame)
                    print("Imagem salva")
                except:
                    print("Não foi possivel salvar imagem. Terminando processo.")

    # Release the capture
    cap.release()
    cv2.destroyAllWindows()
```

```python
def get_informations_hand(index, hand, mark_name, results):
    """
    Retorna um texto com o label (left ou right) e o score para uma mão específica (índice), e
    as coordenadas x e y desta mão para um ponto da mão específico.
    """
    output = None 
    
    # Loop sobre cada mão detectada
    for i, c in enumerate(results.multi_handedness):
        
        # Encontra a mão correta pelo índice
        if c.classification[0].index == index:
            
            # Extrair informações (label e score)
            label = c.classification[0].label
            score = c.classification[0].score
            text = 'Label: {}, Score: {}'.format(label, round(score, 2))
            # print(text)
            
            # Extrair coordenadas de um ponto específico
            # x = results.multi_hand_landmarks[i].landmark[mark_name].x
            # y = results.multi_hand_landmarks[i].landmark[mark_name].y
            # print(x, y)
            
            x = hand.landmark[mark_name].x
            y = hand.landmark[mark_name].y
            # print(x, y)
            
            # Cálculo de coordenadas
            multiply = np.multiply(np.array((x, y)),[640,480])  # TODO: colocar parâmetros da câmera ?
            coords = tuple(multiply.astype(int))
            # print(coords)
            
            output = text, coords
    
    return output
```

```python
cap = cv2.VideoCapture(0)  # Maioria dos casos a câmera do webcam é 0

# Comando 'with' garante que os recursos sejam fechados após a execução do bloco
with mp_hands.Hands(
    min_detection_confidence=0.8,  # Nivel de confiança mínimo para detecção
    min_tracking_confidence=0.5,  # Nível de confiança mínimo para rastreio
    ) as hands:
    
    # Enquanto a câmera está aberta
    while cap.isOpened():
        # Leitura da imagem da câmera
        _, frame = cap.read()  # Frame é lido em formato BGR
        
        # Converte BGR para RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Inverte na horizontal
        image = cv2.flip(image, 1)
        
        # Para aumentar performance flag a imagem, será passada então como referência
        image.flags.writeable = False
        
        # Detecção
        results = hands.process(image)  # Imagem processada em RGB
        # print(results )
        
        # Flag como true
        image.flags.writeable = True
        
        # Converte RGB para BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Mostrando resultados
        if results.multi_hand_landmarks:  # Caso haja algum resultado
            
            # Iteração em cada mão
            for num, hand in enumerate(results.multi_hand_landmarks):
                
                # Desenhar as conexões em cada imagem (frame) da câmera
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                        mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2)
                                        )
                
                # Mostra o label da mão
                if get_informations_hand(num, hand, mp_hands.HandLandmark.WRIST, results):
                    text, coord = get_informations_hand(num, hand, mp_hands.HandLandmark.WRIST, results)
                    
                    # Mostrar na imagem o label e o score
                    cv2.putText(image,  # imagem
                                text,  # texto
                                coord,  # coordenada onde vai ser colocado o texto
                                cv2.FONT_HERSHEY_SIMPLEX,  # fonte
                                0.5,  # escala
                                (255, 255, 255),  # cor
                                1,  # Espessura
                                cv2.LINE_AA)
                    
                    # Mostrar na imagem as coordenadas de um ponto
                    x_ = coord[0]
                    y_ = coord[1] + 30
                    coord_ = tuple((x_, y_))
                    cv2.putText(image,  # imagem
                                str(coord),  # texto
                                coord_,  # coordenada onde vai ser colocado o texto
                                cv2.FONT_HERSHEY_SIMPLEX,  # fonte
                                0.4,  # escala
                                (255, 255, 255),  # cor
                                1,  # Espessura
                                cv2.LINE_AA)
        
        # Mostrar imagem
        cv2.imshow('Hand tracking', image)
        
        # Sair do loop
        if cv2.waitKey(5) & 0xFF == ord('q'):  # Pressionar 'q' para sair
            # Salvar imagem com nome único
            file_name = '{}.jpg'.format(uuid.uuid1())  # Nome único
            cv2.imwrite(output_images_path + file_name, image)  # Salvar frame
            break
            
# Libera recursos alocados
cap.release()
cv2.destroyAllWindows()
```

```python
def str2float(lista):
    for i, value in enumerate(lista):
        lista[i] = float(lista[i])
    return lista
```

```python
def verificar_direcao_mao(hand):
    """
    Verifica se mão está voltada para cima (up) ou para baixo (down).
    """
    # Ponta dos dedos (valor da vertical)
    pos_thumb = hand.landmark[4].y  # Polegar
    pos_index_finger = hand.landmark[8].y  # Indicador
    pos_middle_finger = hand.landmark[12].y  # Médio
    pos_ring_finger = hand.landmark[16].y  # Anelar
    pos_pinky = hand.landmark[20].y  # Mínimo
    
    # Lista dos dedos
    fingers = [pos_thumb, pos_index_finger, pos_middle_finger, pos_ring_finger, pos_pinky]
    
    # Posicao do pulso (referência)
    pos_wrist = hand.landmark[0].y  # Pulso
    
    # Ponto (0,0) é o campo superior esquerda da imagem
    # Se todas as pontas do dedos tiverem valor superior ao valor do pulso, mão está para 'baixo',
    # caso contrário, está para 'cima'
    
    aux = 0  # Variável auxiliar
    for elem  in fingers:
        # print(elem, pos_wrist)
        if elem < pos_wrist:
            aux += 1
    # print(aux)
    
    if aux == 5:  # Todos as pontas dos dedos estão acima do pulso
        return "up"
    else:
        return "down"
```

```python
def verificar_condicao_dedos(finger, hand, hand_direction, img_height):
    """
    Verifica se o dedo está dobrado ou esticado.
    Para verificar se o dedo está esticado ou dobrado é verificado a posição y da ponta e da base do dedo.
    """
    # Caracteristica do dedo
    positions = []
    
    # Determinar conjuntos de pontos para cada dedo
    points = []
    len_finger = 4  # Número de pontos de um dedo
    
    if finger == "thumb":
        init = 1
    elif finger == "index-finger":
        init = 5
    elif finger == "middle-finger":
        init = 9
    elif finger == "ring-finger":
        init = 13
    elif finger == "pinky":
        init = 17
    else:
        raise Exception("Dedo não encontrado")
        
    # Pontos do dedo. 0 para base do dedo, 3 para ponta do dedo
    for i in range(init, init+len_finger):
        points.append(hand.landmark[i].y * img_height)
    # print(f"Pontos do dedo: {points}")
    
    if hand_direction == "up":  # Mão voltada para cima
        if finger == "thumb":  # Polegar
            if points[3] <= points[2]:
                # print(points[2] - points[3])
                # aux = (points[2] - points[3])/(points[2] - points[1])
                # print(f"Teste:{aux}")
                if points[2] - points[3] < 20:  # TODO: adicionar, melhorar lógica
                    positions.append("dobrado")
                else:
                    positions.append("esticado")
            else:  # Polegar dobrado
                positions.append("dobrado")
        else:  # Outros dedos
            if points[3] < points[1]:  # Ponta do dedo com valor menor que base do dedo
            # if points[2] < points[1]:
                positions.append("esticado-vertical")
            else:
                positions.append("dobrado")
    else:  # Mão voltada para baixo
        if finger == "thumb":  # Polegar
            if points[2] <= points[3]:
                if points[3] - points[2] < 20:  # TODO: adicionar, melhorar lógica -> como normalizar ? Pois valor muda se mao está próxima ou longe da tela
                    positions.append("dobrado")
                else:
                    positions.append("esticado")
            else:  # Polegar dobrado
                positions.append("dobrado")
        else:  # Outros
            if points[3] > points[1]:  # Ponta do dedo com valor maior que base do dedo
            # if points[2] > points[1]:
                positions.append("esticado-vertical")
            else:
                positions.append("dobrado")
    return positions
```

```python
def extrair_pontos_dedos(hand, img_height, img_width):
    """
    Cria um dicionário com os valores do pontos dos dedos da mão
    """
    hand_fingers_points = dict()
    
    for finger in HAND_FINGERS:
        if finger == "thumb":
            init = 1
        elif finger == "index-finger":
            init = 5
        elif finger == "middle-finger":
            init = 9
        elif finger == "ring-finger":
            init = 13
        elif finger == "pinky":
            init = 17
        else:
            raise Exception("Dedo não encontrado")
            
        # Pontos do dedo. 0 -> para base do dedo, 3 -> para ponta do dedo
        points_x = []
        points_y = []
        len_finger = 4  # Número de pontos de um dedo
        
        for i in range(init, init+len_finger):
            points_x.append(hand.landmark[i].x * img_width)
            points_y.append(hand.landmark[i].y * img_height)
            
        # Atualiza dicionário
        points = {'x': points_x, 'y': points_y}
        hand_fingers_points.update({finger: points})
    
    return hand_fingers_points
```

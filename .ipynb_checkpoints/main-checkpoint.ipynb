## 1. Acesso a webcam

```python
#
# Cria um objeto para a captura de video do webcam. Para parar o processo pressionar o botão 'q', para salvar o frame (imagem)
# na pasta 'images'.
# Parâmetros: path_images, filename
#

def display_webcam(path_images, filename):
    # Create a video capture object for the camera
    cap = cv2.VideoCapture(0)
    
    while(True):
        # Capture the video frame-by-frame
        _, frame = cap.read()

        # Display the resulting frame
        cv2.imshow('Frame', frame)
        
        # Press 'q' button to quit and
        # 's' button to save the image
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
        elif cv2.waitKey(1) & 0xFF == ord('s'):
            cv2.imwrite(path_images + '\\' + filename, frame)

    # Release the capture
    cap.release()
    cv2.destroyAllWindows()
```

```python
filename = 'letra_c.jpg'
path = os.path.abspath('')
path_images = path + "\\" + 'images'
```

## 2. Media Pipe Hands

### Imagens estáticas

```python
for filename in os.listdir(path_images):
    print(filename)
```

```python
# For static images:
# IMAGE_FILES = []
IMAGE_FILES = [path_images + "\\" + 'letra_a.jpg']

with mp_hands.Hands(
    static_image_mode=True,  # Input é uma imagem
    max_num_hands=2,  # Número máximo de mãos
    min_detection_confidence=0.5) as hands:
  
    for idx, file in enumerate(IMAGE_FILES):
        # Read an image, flip it around y-axis for correct handedness output (see
        # above).
        image = cv2.flip(cv2.imread(file), 1)
        
        # Convert the BGR image to RGB before processing.
        image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        
        # Process the image
        results = hands.process(image_rgb)

        # Print handedness and draw hand landmarks on the image.
        print('Handedness:', results.multi_handedness)
        if not results.multi_hand_landmarks:
          continue

        image_height, image_width, _ = image.shape
        annotated_image = image.copy()

        for hand_landmarks in results.multi_hand_landmarks:
          print('hand_landmarks:', hand_landmarks)
          print(f'Index finger tip coordinates: (',
              f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].x * image_width}, '
              f'{hand_landmarks.landmark[mp_hands.HandLandmark.INDEX_FINGER_TIP].y * image_height})'
          )
          mp_drawing.draw_landmarks(annotated_image, hand_landmarks, mp_hands.HAND_CONNECTIONS)  # Conect the marks

        #cv2.imwrite('/tmp/annotated_image' + str(idx) + '.png', cv2.flip(annotated_image, 1))
        cv2.imwrite(path_images + "\\" + "image_" + str(idx) + '.png', cv2.flip(annotated_image, 1))
```

**MULTI_HANDEDNESS**<br>
Coleção das mãos detectadas (ou seja, é uma mão esquerda ou direita). Cada mão é composta por um _label_ e _score_. O label é uma string com o valor "Esquerda" ou "Direita". O score é a probabilidade estimada da mão prevista e é sempre maior ou igual a 0.5 (e a mão oposta tem uma probabilidade estimada de 1-score).

Note-se que a mão é determinada assumindo que a imagem de entrada é espelhada, ou seja, tirada com uma câmara virada para a frente com as imagens viradas na horizontal. Se não for este o caso, por favor trocar a saída da mão na aplicação.

```python
results.multi_handedness
```

**MULTI_HAND_LANDMARKS** <br>
Conjunto de mãos detectadas, onde cada mão é representada como uma lista de _21 marcos_ de mão e cada marco de referência (ponto de referência) é composto por x, y e z. Desses, x e y são normalizados para [0.0, 1.0] pela largura e altura da imagem, respectivamente. Já z representa a profundidade do marco de referência, sendo a profundidade no pulso a origem (referência). Quanto menor o valor de z, mais próximo o marco de referência está da câmera. A magnitude de z usa aproximadamente a mesma escala que x.

### Imagem da câmera webcam 

```python
# For webcam input:
cap = cv2.VideoCapture(0)
with mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5) as hands:
  while cap.isOpened():
    success, image = cap.read()
    if not success:
      print("Ignoring empty camera frame.")
      # If loading a video, use 'break' instead of 'continue'.
      continue

    # Flip the image horizontally for a later selfie-view display, and convert
    # the BGR image to RGB.
    image = cv2.cvtColor(cv2.flip(image, 1), cv2.COLOR_BGR2RGB)
    # To improve performance, optionally mark the image as not writeable to
    # pass by reference.
    image.flags.writeable = False
    results = hands.process(image)
    print(results)

    # Draw the hand annotations on the image.
    image.flags.writeable = True
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
        mp_drawing.draw_landmarks(
            image, hand_landmarks, mp_hands.HAND_CONNECTIONS)
    cv2.imshow('MediaPipe Hands', image)
    if cv2.waitKey(5) & 0xFF == 27:
      break
cap.release()
```
